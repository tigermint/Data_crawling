```python
url = "https://www.youtube.com/watch?v=4TWR90KJl84"
# í¬ë¡¬ ë¸Œë¼ìš°ì €ì— ì ‘ì†í•  í˜ì´ì§€ë¥¼ ì„¤ì •í•œë‹¤.
driver.get(url)
```

```python
# execute_script("jsì‘ì„±ê°€ëŠ¥") : ìŠ¤í¬ë¡¤ í•˜ëŠ” ë©”ì„œë“œ 
# get scroll height
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
```

```python
# ëŒ“ê¸€ì„ ìŠ¤í¬ë¡¤ì„ ëª‡ ë²ˆ ë‚´ë ¤ ëŒ“ê¸€ì„ ê°€ì ¸ì˜¬ê¹Œ?
for i in range(5):
    # í™”ë©´ìƒ ìŠ¤í¬ë¡¤ ìœ„ì¹˜ì´ë™ : scrollTo(x,y)
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight)")
    time.sleep(3)
    # ë‹¤ìŒ ë„˜ì–´ê°€ëŠ” ìƒˆë¡œìš´ í˜ì´ì§€ ë†’ì´
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    if new_page_height == last_page_height:
        break
    last_page_height = new_page_height
```

```python
html_source = driver.page_source
```

```python
soup = BeautifulSoup(html_source, "html.parser")
```

```python
youtube_user_IDs = soup.select("#author-text > span")
youtube_comments = soup.select("#content-text")
print(youtube_user_IDs, youtube_comments)
```

```python
youtube_user_ID = soup.select("#author-text > span")[0].text.strip()
youtube_comment = soup.select("#content-text")[0].text.strip()
print(youtube_user_ID,"\n", youtube_comment)
```

```python
str_youtube_userIDs = []
str_youtube_comments = []

for i in range(len(youtube_user_IDs)):
    str_tmp = str(soup.select("#author-text > span")[i].text.strip())
    print(str_tmp)
    str_youtube_userIDs.append(str_tmp)
    
    str_tmp = str(soup.select("#content-text")[i].text.strip())
    print(str_tmp)
    str_youtube_comments.append(str_tmp)
```

```python
import pandas as pd

pd_data = {"ID": str_youtube_userIDs, "comments": str_youtube_comments}
youtube_df = pd.DataFrame(pd_data)
```

```python
# 크롤링 결과 중 해시태그 데이터 불러오기
import pandas as pd
raw_total = pd.read_excel("./files/1_crawling_daegucafe_500.xlsx")
raw_total["tags"][:10] # tags colum 0 ~ 9 까지 불러온다
# tags 10개만 가져오기
```

```python
# 해시태그 통합 저장하기
tags_total = [] # list 형태

for tags in raw_total["tags"]: # raw_total["tags"] 안에 들어있는 tags수 만큼 for문 돌림, tags는 lis
    tags_list = tags[2:-2].split("', '") #???????????????????????????
    # tags_list에 각 태그의 2 ~ -2까지 ', '이 간격마다 잘라준다 [start:stop:step], stop은 exclusive
    # split 함수: 문자열.split("구분자")
    for tag in tags_list:
         tags_total.append(tag) # 각각의 구분된 tags_list를 빈 tags_total이란 리스트에 넣어준다
tags_total
```

```python
# 빈도수 집계하기(Counter)
from collections import Counter # collections 모듈에서 Counter class 들고 온다 -> 얼마나 사용되었는지
tag_counts = Counter(tags_total)
```

```python
# 가장 많이 사용된 해시태그 살펴보기
tag_counts.most_common(50) # 데이터의 개수가 많은 순으로 정렬된 배열을 리턴하는 메서드
```

```python
# 데이터 정제하기
STOPWORDS = ['', '#일상','#맞팔', '#좋아요','#daily', '#좋아요반사', '#좋반', '#팔로우', '#맞팔', '#selfie', '#일상', 
            '#ootd', '#소통', '#오오티디', '#선팔','#맞팔', '#인친']
tag_total_selected = []
for tag in tags_total:
    if tag not in STOPWORDS:
        tag_total_selected.append(tag)
        
tag_counts_selected = Counter(tag_total_selected)
tag_counts_selected.most_common(50)
```

```python
# 시각화 라이브러리 호출 -> 막대 차트로 해시태그 살펴보기
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import sys

if sys.platform in ["win32", "win64"]:
    font_name = "malgun gothic"
elif sys.platform == "dawin":
    font_name = "AppleGothic"
    
rc("font", family = font_name)
```

```python
# 데이터 준비하기
tag_counts_df = pd.DataFrame(tag_counts_selected.most_common(30))
tag_counts_df.columns = ["tags", "counts"]
```

```python
# 막대 차트 그리기
plt.figure(figsize=(10,8))
# plt.figure()
sns.barplot(x = "counts", y = "tags", data = tag_counts_df)
plt.show()
```

```python
# 워드클라우드 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import platform

if platform.system() == "Windows": #윈도우
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == "Darwin": # 맥
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
# 워드클라우드 만들기
wordcloud = WordCloud(font_path = font_path, 
                      background_color = "white",
                      max_words = 100,
                      relative_scaling=0.3,
                      width = 800,
                      height = 400,
                      ).generate_from_frequencies(tag_counts_selected)
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("./files/tag_wordcloud.png")
```

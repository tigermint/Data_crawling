```python
# 크롤링 데이터 불러오기

import pandas as pd
raw_total = pd.read_excel("./files/1_crawling_daegucafe_500.xlsx")# pd.read로 엑셀파일 읽는다
raw_total.head()
```

```python
# 추출 단어 포함하는 데이터 선책하기
select_word = "수성못"

check_list = []
for content in raw_total["content"]:# column이 content인 것들만큼 for문 돌림
    if select_word in content: # content 중 select_word 인게 있으면 빈 깡통 check_list에 append한다
        check_list.append(True)
    else:
        check_list.append(False)
select_df = raw_total[check_list] # raw_total[check_list]를 select_df라 한다
select_df.head()
```

```python
# 선택 데이터 확인하기
for i in select_df.index:
    print(select_df.loc[i, "content"])
    print("-" * 50)
```

```python
# 선택 데이터 저장하기
fpath = f"./files/deledcted_data{select_word}.xlsx"
select_df.to_excel(fpath)
```

```python
# 여러개 단어 선택/추출/저장하기
select_word_list = ["수성못", "앞산", "교동", "삼덕동"]
for select_word in select_word_list:
    check_list = []
    for content in raw_total["content"]: # 맨위에와 같은데 이건 각각 select_word_list의 단어들 별로 content 추출
        if select_word in content:
            check_list.append(True)
        else:
            check_list.append(False)

    select_df = raw_total[check_list] # 각각 select_df를 만들고
        
    fpath = f"./files/deledcted_data{select_word}.xlsx"# 파일에 저장한다
    select_df.to_excel(fpath)
```

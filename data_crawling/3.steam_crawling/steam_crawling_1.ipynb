```python
# 파이썬 크롬 제어하기 - Chromedriver
from selenium import webdriver

driver = webdriver.Chrome("C:/Users/SAMSUNG/Desktop/2021KnuPython/step2/chromedriver")
```

```python
import time

# steam 접속하기
driver.get("https://store.steampowered.com/search/?filter=topsellers")
time.sleep(2)
```

```python
# 웹 페이지의 html 다운로드
html = driver.page_source
html
```

```python
from bs4 import BeautifulSoup

# html 파일의 문자열 정보를 html 형식에 맞게 해석
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
# 게임 이름 
name = soup.select("div.search_name > span")[0].text
#soup.select('원하는 정보')  # select('원하는 정보') -->  단 하나만 있더라도, 복수 가능한 형태로 되어있음
name
```

```python
# 게임 출시일 
released = soup.select("div.search_released")[0].text
released
```

```python
# 게임 가격
price = soup.select("div.search_price")[1].text.strip().split("₩ ")[1]
# strip() : 기존 text에서 정리되지 않은 걸 정리해주는 함수 -> 양쪽 공백 지우기
price
```

```python
# 게임 전체 목록 가져오기
games = soup.select("a.search_result_row")
games
```

```python
# 하나씩 출력하기
for game in games:
    print(game)
    time.sleep(3)
```

```python
# 게임 이름
print(games[0].select("div.search_name > span")[0].text)

# 게임 출시일 
print(games[0].select("div.search_released")[0].text)

# 게임 가격
print(games[0].select("div.search_price")[0].text.strip().split("₩ ")[1])
```

```python
for game in games:
    # 게임 이름
    name = game.select("div.search_name > span")[0].text

    # 게임 출시일 
    released = game.select("div.search_released")[0].text

    # 게임 가격
    price = game.select("div.search_price")[0].text.strip().split("₩ ")[1]
    print(name, released, price)    
```

```python
# 추천 게임 목록 URL 가져오기
top_sellers_games = soup.select("#search_resultsRows > a")
top_sellers_game_urls = []

for top_sellers_game in top_sellers_games:
    top_sellers_game_urls.append(top_sellers_game["href"])
top_sellers_game_urls
```

```python
classification = pd.read_excel("./files/daegu_fashion_classification.xlsx")
classification
```

```python
# 병합 merge(df1, df2, on = "key") 
# 데이터 옆으로 통합하기(left)
sample_1_code = pd.merge(left = sample_1, # 왼쪽엔 sample_1이 들어간다 
                        right = classification, # 오른쪽엔 classification이 들어간다
                        how = "left", # 왼쪽에 있는 df1이 기준, 중복된 column 제외 옆으로 달라붙게 된다
                        left_on = "MTI Code",
                        right_on = "MTI Code") # 이거 기준으로 공통된 인자가 df1, df2 에서 Mti Code값이 같은 것 기준으로 m * n형태로 나타나게 된다
sample_1_code
```

```python
# 데이터 옆으로 통합하기(inner)
sample_1_code = pd.merge(left = sample_1,  
                        right = classification, 
                        how = "inner",  
                        left_on = "MTI Code",
                        right_on = "MTI Code") 
sample_1_code
```

```python
# 데이터 통합하기
sample_2 = pd.read_excel("./files/daegu_fashion_2.xlsx",
                        header = 1,
                        skipfooter = 2,
                        usecols = "A:C")
sample_2["기준년"] = 2019
sample_2_code = pd.merge(left = sample_2,  
                        right = classification, 
                        how = "left",  
                        left_on = "MTI Code",
                        right_on = "MTI Code")
sample_2_code
```

```python
sample = sample_1_code.append(sample_2_code, ignore_index = True)
sample
```

```python
sample.to_excel("./files/dague_fashion.xlsx")
```

```python
sample.to_excel("./files/dague_fashion_index_false.xlsx", index = False)
```

```python
# 데이터 집계하기
sample_pivot = sample.pivot_table(values = "수출액",
                                 index = "구분",
                                 columns = "항목",
                                 aggfunc = "mean")
sample_pivot
```
